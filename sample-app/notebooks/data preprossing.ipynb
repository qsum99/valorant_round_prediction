```python
pre_df=df.copy()
live_df=df1.copy()
```

```python
pre_df_clean=pre_df.dropna(subset=["winner"]).copy()
live_df_clean=live_df.dropna(subset=["winner"]).copy()
```

```python
live_df_clean.info()
```

```python
## lets do some graps
# Graph 1: Class Balance (Pre-Round)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(
    data=pre_df_clean, 
    x=pre_df['winner'], 
    hue=pre_df['winner'], 
    legend=False, 
    palette='viridis'
)
plt.title('Pre-Round: Class Balance (Winner)', fontsize=16)
plt.xlabel('Winner', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()

```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=pre_df_clean, x='map', order=pre_df_clean['map'].value_counts().index, palette='mako')
plt.title('Pre-Round: Map Distribution', fontsize=16)
plt.xlabel('Map', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=pre_df_clean, x='score_diff', hue='winner', element="step", stat="density", common_norm=False, palette='rocket')
plt.title('Pre-Round: Score Difference Distribution by Winner', fontsize=16)
plt.xlabel('Score Difference (Local - Opponent)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=pre_df_clean, x='winner', y='economy_diff', palette='viridis')
plt.title('Pre-Round: Economy Difference by Winner', fontsize=16)
plt.xlabel('Winner', fontsize=12)
plt.ylabel('Economy Difference (Local - Opponent)', fontsize=12)
plt.tight_layout()
```
