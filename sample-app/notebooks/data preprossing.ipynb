```python
pre_df=df.copy()
live_df=df1.copy()
```

```python
pre_df_clean=pre_df.dropna(subset=["winner"]).copy()
live_df_clean=live_df.dropna(subset=["winner"]).copy()
```

```python
live_df_clean.info()
```

```python
## lets do some graps
# Graph 1: Class Balance (Pre-Round)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(
    data=pre_df_clean, 
    x=pre_df['winner'], 
    hue=pre_df['winner'], 
    legend=False, 
    palette='viridis'
)
plt.title('Pre-Round: Class Balance (Winner)', fontsize=16)
plt.xlabel('Winner', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()

```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=pre_df_clean, x='map', order=pre_df_clean['map'].value_counts().index, palette='mako')
plt.title('Pre-Round: Map Distribution', fontsize=16)
plt.xlabel('Map', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=pre_df_clean, x='score_diff', hue='winner', element="step", stat="density", common_norm=False, palette='rocket')
plt.title('Pre-Round: Score Difference Distribution by Winner', fontsize=16)
plt.xlabel('Score Difference (Local - Opponent)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=pre_df_clean, x='winner', y='economy_diff', palette='viridis')
plt.title('Pre-Round: Economy Difference by Winner', fontsize=16)
plt.xlabel('Winner', fontsize=12)
plt.ylabel('Economy Difference (Local - Opponent)', fontsize=12)
plt.tight_layout()
```

```python
## lets do grap for the live_pre data
live_numeric = live_df_clean.select_dtypes(include=[np.number])
live_numeric_subset = live_numeric.drop(columns=['round_number', 'kill_index'], errors='ignore')
plt.figure(figsize=(12, 10))
corr_matrix = live_numeric_subset.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Live-Round: Feature Correlation Heatmap', fontsize=16)
plt.tight_layout()
```

```python
avg_alive_diff_per_round = live_df_clean.groupby(['round_number', 'winner'])['alive_diff'].mean().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(data=avg_alive_diff_per_round, x='round_number', y='alive_diff', hue='winner', marker='o', palette='Set1')
plt.title('Live-Round: Average Alive Difference by Round Number and Winner', fontsize=16)
plt.xlabel('Round Number', fontsize=12)
plt.ylabel('Average Alive Difference (Local - Opponent)', fontsize=12)
plt.axhline(0, color='gray', linestyle='--')
plt.tight_layout()
```
