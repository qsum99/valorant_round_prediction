```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("pre_round.csv")
```

```python
df.describe()
```

```python
df1=pd.read_csv("live_round.csv")
```

```python
df1.describe()
```

```python
df.info()
```

```python
df1.info()
```

```python
pre_df=df.copy()
live_df=df1.copy()
```

```python
pre_df_clean=pre_df.dropna(subset=["winner"]).copy()
live_df_clean=live_df.dropna(subset=["winner"]).copy()
```

```python
live_df_clean.info()
```

```python
## lets do some graps
# Graph 1: Class Balance (Pre-Round)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(
    data=pre_df_clean, 
    x=pre_df['winner'], 
    hue=pre_df['winner'], 
    legend=False, 
    palette='viridis'
)
plt.title('Pre-Round: Class Balance (Winner)', fontsize=16)
plt.xlabel('Winner', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.close()
```
